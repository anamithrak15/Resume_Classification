# Cleaning data
- URLs
- Hashtags
- Mentions
- Special Characters
- Punctuations


```python
import re 
```

```python
def clean_resume(text):
    cleantxt=re.sub('http\S+\s',' ',text)
    cleantxt=re.sub('@\S+',' ',cleantxt)
    cleantxt=re.sub('#\S+\s',' ', cleantxt)
    cleantxt=re.sub('RT|cc',' ',cleantxt)
    cleantxt=re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""),' ',cleantxt)
    cleantxt=re.sub(r'[^\x00-\x7f]',' ',cleantxt)
    cleantxt=re.sub('\s+',' ',cleantxt)
    return cleantxt
```

```python
df['Resume']=df['Resume'].apply(lambda x: clean_resume(x))
```

# KNN
- KNN is based on the idea that similar data points tend to belong to the same class. Given a new data point, KNN finds the K nearest data points in the training set and assigns the majority class among these neighbors to the new data point.

- Train : KNN does not explicitly "train" a model in the traditional sense. Instead, it memorizes the entire training dataset.

- Prediction : To predict the class of a new data point:
    - Calculate the distance between the new data point and each point in the training set using a distance metric (e.g., Euclidean distance).
    - Select the K nearest neighbors (data points) based on the calculated distances.
    - Assign the class label that occurs most frequently among the K nearest neighbors to the new data point.

```python
clf=KNeighborsClassifier()
clf.fit(X_train, y_train)
ypred = clf.predict(X_test)
print(ypred)
```

# How OneVsRestClassifier works:

- Binary Classification: For each class in the multiclass problem, a separate binary classifier is trained. Each binary classifier is trained to distinguish between one class and the rest of the classes. For example, if you have 3 classes (A, B, and C), you would train 3 binary classifiers:

    - Classifier 1: Class A vs. not Class A (i.e., classes B and C)
    - Classifier 2: Class B vs. not Class B (i.e., classes A and C)
    - Classifier 3: Class C vs. not Class C (i.e., classes A and B)
- Combining Predictions: During prediction, each binary classifier makes a prediction for its respective class. The final prediction for a given instance is the combination of predictions from all binary classifiers. This typically involves taking the class with the highest confidence score among all binary classifiers. 

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
```

```python
clf=OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train, y_train)
ypred = clf.predict(X_test)
print(ypred)
```

```python
print(accuracy_score(y_test, ypred))
```
